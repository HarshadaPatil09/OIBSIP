```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.express as px
```

```python
df=pd.read_csv('/content/Unemployment in India.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.columns
```

```python
df.rename(columns={'Region':'State',' Date':'Date',' Frequency':'Frequency',' Estimated Unemployment Rate (%)':'Unemployment Rate',' Estimated Employed':'Employed',' Estimated Labour Participation Rate (%)':'Labour Rate','Area':'Area'},inplace=True)
```

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.replace(np.NaN,df.mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(['Date'],axis=1)
```

```python
df['State'].unique()
```

```python
plt.figure(figsize=(10, 12))
plt.title("Unemployment Rate")
sns.barplot(x="Unemployment Rate", y="State", data=df)
plt.show()
```

```python
heat_maps = df[['Unemployment Rate','Employed', 'Labour Rate']]

heat_maps = heat_maps.corr()

plt.figure(figsize=(10,6))
sns.set_context('notebook')
sns.heatmap(heat_maps, annot=True,cmap='rainbow')
```

```python
df.groupby(['State'])[['Unemployment Rate']].mean().plot(kind='bar')
```

```python
df.groupby(['State'])[['Employed']].mean().plot(kind='bar')
```

```python
df.groupby(['Area'])[['Unemployment Rate']].mean().plot(kind='bar')
```

```python
df.groupby(['Area'])[['Employed']].mean().plot(kind='bar',color='red')
```

```python
df['Area'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
plt.title("Estimated Unemployment Rate")
sns.histplot(x= 'Unemployment Rate',hue='Area',data=df)
```

```python
sns.histplot(x='Unemployment Rate',hue='State',data=df)
plt.title('Unemployment Rate')
```
